# **Exercícios**

## 1\. Pinguins

Neste exercício, vamos utilizar uma base de dados com informações sobre penguins. A idéia é preparar a base de dados para prever a espécie do penguin (variável resposta) baseado em suas características físicas e geográficas (variáveis preditivas).

```
import numpy as np
import pandas as pd
import seaborn as sns
import re
```

```
data = sns.load_dataset('penguins')
```

```
data.head()
```

### **1.1. Valores nulos**

A base de dados possui valores faltantes, utilize os conceitos da aula para trata-los.

```
# resposta da questão 1.1
data[['species', 'island', 'sex']].isnull().any()
df_limpo = data.dropna(subset = 'sex')
df_limpo[['species', 'island', 'sex']].isnull().any()
df_limpo[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']].isnull().any()
```

### **1.2. Variáveis numéricas**

Identifique as variáveis numéricas e crie uma nova coluna **padronizando** seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_std*".

> **Nota**: Você não deve tratar a variável resposta.

```
# resposta da questão 1.2
desvio_df = df_limpo[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']]

#Padronizando padronização em (sigma = (x - média) / desvio padrão)
desvio_df = desvio_df.apply(lambda x: (x - np.mean(x)) / np.std(x))

#Utilizando métodos para conferir a padronização
desvio_df.describe()
for coluna in desvio_df:
  desvio_df.rename(columns = {coluna: coluna + '_std'}, inplace = True)

df_limpo = pd.merge(left  = df_limpo, right = desvio_df, how = 'inner', right_index = True, left_index = True)
```

### **1.3. Variáveis categóricas**

Identifique as variáveis categóricas nominais e ordinais, crie uma nova coluna aplicando a técnica correta de conversão a seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_nom*" ou "*_ord*".

> **Nota**: Você não deve tratar a variável resposta.

```
# resposta da questão 1.3

#Criação de um DataFrame em que seus valores serão convertidos para valores binários
aux_df = pd.get_dummies(data = df_limpo[['island', 'sex']])

# Alteração do nome das colunas
for coluna in aux_df:
  aux_df.rename(columns = {coluna: coluna + '_nom'}, inplace = True)

df_limpo = pd.merge(left = df_limpo, right = aux_df, how = 'inner', right_index = True, left_index = True)
df_limpo
```

### **1.4. Limpeza**

Descarte as colunas originais e mantenha apenas a variável resposta e as variáveis preditivas com o sufixo *_std*", *_nom*" e "*_ord*".

```
# resposta da questão 1.4
columns_list = ['species']

#As colunas com nomes em _ std ou _nom serão adicionadas a uma lista com seus respectivos nomes de colunas
for coluna in df_limpo:
  if re.findall(".*_std.*|.*_nom.*", coluna):
    columns_list.append(coluna)

#Criação do DataFrame Penguins com as mudanças realizadas
penguins_df = df_limpo[columns_list]
penguins_df
```
