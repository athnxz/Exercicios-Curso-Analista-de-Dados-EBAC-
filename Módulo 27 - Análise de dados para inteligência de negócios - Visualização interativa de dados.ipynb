### **2.2. Transformação**

Tranforme o texto da coluna `month` para sua representação numérica, exemplo: `Jan` para `1` e `Dec` para `12`.

```
# resolução
month = []
month = flights['month']
month = month.replace('Jan', '01').replace('Feb', '02').replace('Mar', '03').replace('Apr', '04').replace('May', '05').replace('Jun', '06').replace('Jul', '07').replace('Aug', '08').replace('Sep', '09').replace('Oct', '10').replace('Nov', '11').replace('Dec', '12')
df_month = flights.drop('month', axis=1)
df_month['month'] = month
df_month
```

Crie uma chave temporal `year-month` no formato `YYYY-MM` através da concatenação das colunas `year` e `month`, exemplo: `1949-01`.

```
# resolução
month
year = []
year = flights['year']
year_month = []
year_month = [None] * len(month)
i=0
for i in range(len(month)):
    aux = month[i]
    aux1 = year[i]
    aux2 = str(aux)+"-"+str(aux1)
    year_month[i] = aux2
    i=i+1

df_month['year-month'] = year_month
df_month
```

Reordene as colunas da seguinte forma: `year-month`, `year`, `month` e `passengers`.

```
# resolução
df = df_month[['year-month', 'year', 'month', 'passengers']]
df
```

### **2.3. Carregamento**

Persista o DataFrame em um arquivo com o nome `flights.csv`, no formato CSV. Garanta que o índice será descartado.

```
# resolução
df.to_csv('flights.csv', index = False)
```
